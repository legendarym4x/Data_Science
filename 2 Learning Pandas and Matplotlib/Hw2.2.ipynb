### Analyze the 2017_jun_final.csv file. The file contains the results of the developer survey in June 2017.

### 1. Read the file 2017_jun_final.csv using the read_csv method:

```python
import pandas as pd

jun = pd.read_csv("2017_jun_final.csv")
jun
```

### 2. Read the resulting table using the head method 'head':

```python
jun.head()
```

### 3. Determine the size of the table using the shape method:

```python
print("The number of rows and columns:", jun.shape)
```

### 4. Define the types of all columns using 'dataframe.dtypes':

```python
print(jun.dtypes)
```

### 5. Calculate the percentage of blanks in each column (use isnull and sum methods):

```python
missing_values = jun.isnull().sum()
print(missing_values)
```

### 6. Remove all blank columns except the "Programming Language" column:

```python
remove_blanks = jun.drop(['Спеціалізація', 'Університет', 'Предметна область', 'Валюта', 'cls'], axis=1)

remove_blanks.head()
```

### 7. Count again what proportion of gaps are in each column and make sure that only the column "Мова програмування" remains:

```python
print('Our table after removing unnecessary data:\n', remove_blanks.isnull().sum())
```

### 8. Delete all rows in the source table using the 'dropna' method:

```python
remove_all_blanks = remove_blanks.dropna()

print('Our table after removing unnecessary data:\n', remove_all_blanks.isnull().sum())
```

### 9. Define the new size of the table using the 'shape' method:

```python
print("The number of rows and columns:", remove_all_blanks.shape)
```

### 10. Create a new python_data table, which will contain only rows with specialists who specified the Python programming language:

```python
python_data = remove_all_blanks.loc[remove_all_blanks['Мова програмування'] == 'Python']
python_data
```

### 11. Determine the size of the python_data table using the 'shape' method:

```python
print("The number of rows and columns:", python_data.shape)
```

### 12. Using the 'groupby' method, group by the Position column:

```python
group_by_salary = python_data.groupby('Посада')
print(group_by_salary)
```

### 12. Create a new DataFrame where for the data grouped by the Position column, aggregate the data using the agg method and find the min and max values in the Salary.monthly column:


```python
group_by_salary = python_data.groupby('Посада')
aggregated_data = group_by_salary.agg({'Зарплата на місяць': ['min', 'max']})
print(aggregated_data)
```

### 13. Create a function fill_avg_salary that will return the average salary for a month. Use it for the apply method and create a new column "avg":

```python
def fill_avg_salary(group):
    avg_salary = group['Зарплата на місяць'].mean()
    group['avg'] = avg_salary
    return group

avg = group_by_salary.apply(fill_avg_salary)
avg
```

### 14. Create descriptive statistics using the describe method for the new column:

```python
avg_stats = avg['avg'].describe()
avg_stats
```

### 15. Save the resulting table in a CSV file:

```python
avg.to_csv('avg_statistics.csv')
```

### 16. We build random graphs:

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

exp_salary = avg[['exp', 'salary']]
plt.figure(figsize=(10, 6))
sns.lineplot(x='exp', y='salary', data=exp_salary)
plt.title('Залежність зарплати від стажу', color='blue')
plt.xlabel('Стаж (роки)', color='blue')
plt.ylabel('Зарплата за місяць', color='blue')
plt.grid()
plt.show()
```

```python
table_avg = avg[['Посада', 'avg']]
table_avg = table_avg.drop_duplicates(subset='Посада').set_index('Посада')

sns.set_style("whitegrid")

plt.figure(figsize=(10, 4))
sns.barplot(x='Посада', y='avg', data=table_avg, palette='magma', hue='Посада', dodge=False)
plt.title('Середній рівень зарплати за посадою')
plt.xticks(rotation=45, ha='right')
plt.show()
```
